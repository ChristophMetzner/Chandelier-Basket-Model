### Plotting functions and plotting related parameters


```python
s               = 2**13
sim_time        = 500             
dt              = float(sim_time)/float(s)

def plotPSDs(freqs,p20,p24,p28,xlabel,ylabel,savefig,filename,fontsizes):
    fig = plt.figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    
    ax.plot(freqs,p20,'k',linewidth=4)
    ax.plot(freqs,p24,'r',linewidth=4)
    ax.plot(freqs,p28,'g',linewidth=4)
    ax.axis(xmin=0, xmax=55)
    ax.set_xlabel(xlabel,fontsize=fontsizes[0])
    ax.set_ylabel(ylabel,fontsize=fontsizes[0])
    
    xticks=ax.xaxis.get_ticklabels()
    xticks[2].set_weight('bold')
    xticks[3].set_weight('bold')
    xticks[4].set_weight('bold')
    
    plt.setp(ax.get_xticklabels(),visible=True,fontsize=fontsizes[2])
    plt.setp(ax.get_yticklabels(),visible=True,fontsize=fontsizes[2])
    
    if savefig:
        plt.savefig(filename+'.eps',dpi=600)

def plotPSDzoom1(freqs,p20,p24,p28,ub20,lb20,ub24,lb24,ub28,lb28,xlabel,ylabel,savefig,filename,fontsizes):
    fig = plt.figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    
    ax.plot(freqs,p20,'k',linewidth=4)
    ax.plot(freqs,p24,'r',linewidth=4)
    ax.plot(freqs,p28,'g',linewidth=4)
    ax.axis(xmin=38, xmax=42)
    ax.set_xlabel(xlabel,fontsize=fontsizes[0])
    ax.set_ylabel(ylabel,fontsize=fontsizes[0])
    
    xticks=ax.xaxis.get_ticklabels()
    xticks[4].set_weight('bold')
    
    plt.setp(ax.get_xticklabels(),visible=True,fontsize=fontsizes[2])
    plt.setp(ax.get_yticklabels(),visible=True,fontsize=fontsizes[2])
    
    
    ax.fill_between(freqs, ub20, lb20, color='k', alpha=.5)
    ax.fill_between(freqs, ub24, lb24, color='r', alpha=.5)
    ax.fill_between(freqs, ub28, lb28, color='g', alpha=.5)
    
    if savefig:
        plt.savefig(filename+'.eps',dpi=600) 
        
def plotPSDzoom2(freqs,p20,p24,p28,ub20,lb20,ub24,lb24,ub28,lb28,xlabel,ylabel,savefig,filename,fontsizes):
    fig = plt.figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    
    ax.plot(freqs,p20,'k',linewidth=4)
    ax.plot(freqs,p24,'r',linewidth=4)
    ax.plot(freqs,p28,'g',linewidth=4)
    ax.axis(xmin=18, xmax=22)
    ax.set_xlabel(xlabel,fontsize=fontsizes[0])
    ax.set_ylabel(ylabel,fontsize=fontsizes[0])
    
    xticks=ax.xaxis.get_ticklabels()
    xticks[4].set_weight('bold')
    
    plt.setp(ax.get_xticklabels(),visible=True,fontsize=fontsizes[2])
    plt.setp(ax.get_yticklabels(),visible=True,fontsize=fontsizes[2])
    
    
    ax.fill_between(freqs, ub20, lb20, color='k', alpha=.5)
    ax.fill_between(freqs, ub24, lb24, color='r', alpha=.5)
    ax.fill_between(freqs, ub28, lb28, color='g', alpha=.5)
    
    if savefig:
        plt.savefig(filename+'.eps',dpi=600)
        
def plotPSDzoom3(freqs,p20,p24,p28,ub20,lb20,ub24,lb24,ub28,lb28,xlabel,ylabel,savefig,filename,fontsizes):
    fig = plt.figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    
    ax.plot(freqs,p20,'k',linewidth=4)
    ax.plot(freqs,p24,'r',linewidth=4)
    ax.plot(freqs,p28,'g',linewidth=4)
    ax.axis(xmin=28, xmax=32)
    ax.set_xlabel(xlabel,fontsize=fontsizes[0])
    ax.set_ylabel(ylabel,fontsize=fontsizes[0])
    
    xticks=ax.xaxis.get_ticklabels()
    xticks[4].set_weight('bold')
    
    plt.setp(ax.get_xticklabels(),visible=True,fontsize=fontsizes[2])
    plt.setp(ax.get_yticklabels(),visible=True,fontsize=fontsizes[2])
    
    
    ax.fill_between(freqs, ub20, lb20, color='k', alpha=.5)
    ax.fill_between(freqs, ub24, lb24, color='r', alpha=.5)
    ax.fill_between(freqs, ub28, lb28, color='g', alpha=.5)
    
    if savefig:
        plt.savefig(filename+'.eps',dpi=600) 
        
def barplot2(d,d2,xlabel,ylabel,xticklabels,savefig,filename,fontsizes):
    fig = plt.figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    
    
    ind = np.arange(len(d))
    width = 0.25
    
    rects1 = ax.bar(ind,d,width, color='k')
    rects2 = ax.bar(ind+width,d2,width, color='0.5')
    
    ax.set_ylabel(ylabel,fontsize=fontsizes[0])
    ax.set_xlabel(xlabel,fontsize=fontsizes[0])
    ax.set_xticks(ind)
    ax.set_xticklabels(xticklabels)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(fontsizes[1]) 
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(fontsizes[1]) 
    
    if savefig:
        plt.savefig(filename,dpi=600)
```

### Figure 4040 and 2040 measures - 10% ChCs

```python
d40_20 = data_20[2][2][1]
d40_24 = data_24[2][2][1]
d40_28 = data_28[2][2][1]

p40_20,freqs = calcPowerSpectrum(d40_20,dt,sim_time)
p40_24,freqs = calcPowerSpectrum(d40_24,dt,sim_time)
p40_28,freqs = calcPowerSpectrum(d40_28,dt,sim_time)

p4040_20 = np.sum(p40_20[19:22])
p4040_24 = np.sum(p40_24[19:22])
p4040_28 = np.sum(p40_28[19:22])

p2040_20 = np.sum(p40_20[9:12])
p2040_24 = np.sum(p40_24[9:12])
p2040_28 = np.sum(p40_28[9:12])
```

```python
psd20_40hzstim = np.zeros((20,4096))
psd20_30hzstim = np.zeros((20,4096))
psd20_20hzstim = np.zeros((20,4096))

psd24_40hzstim = np.zeros((20,4096))
psd24_30hzstim = np.zeros((20,4096))
psd24_20hzstim = np.zeros((20,4096))

psd28_40hzstim = np.zeros((20,4096))
psd28_30hzstim = np.zeros((20,4096))
psd28_20hzstim = np.zeros((20,4096))

for j in range(20):
    psd20_40hzstim[j],freqs = calcPowerSpectrum(sdata20_40hzstim[j],dt,sim_time)
    psd20_30hzstim[j],freqs = calcPowerSpectrum(sdata20_30hzstim[j],dt,sim_time)
    psd20_20hzstim[j],freqs = calcPowerSpectrum(sdata20_20hzstim[j],dt,sim_time)

    psd24_40hzstim[j],freqs = calcPowerSpectrum(sdata24_40hzstim[j],dt,sim_time)
    psd24_30hzstim[j],freqs = calcPowerSpectrum(sdata24_30hzstim[j],dt,sim_time)
    psd24_20hzstim[j],freqs = calcPowerSpectrum(sdata24_20hzstim[j],dt,sim_time)
    
    psd28_40hzstim[j],freqs = calcPowerSpectrum(sdata28_40hzstim[j],dt,sim_time)
    psd28_30hzstim[j],freqs = calcPowerSpectrum(sdata28_30hzstim[j],dt,sim_time)
    psd28_20hzstim[j],freqs = calcPowerSpectrum(sdata28_20hzstim[j],dt,sim_time)

    
std20_40hzstim = np.std(psd20_40hzstim, axis=0)  
std20_30hzstim = np.std(psd20_30hzstim, axis=0)  
std20_20hzstim = np.std(psd20_20hzstim, axis=0)  

std24_40hzstim = np.std(psd24_40hzstim, axis=0)  
std24_30hzstim = np.std(psd24_30hzstim, axis=0)  
std24_20hzstim = np.std(psd24_20hzstim, axis=0)

std28_40hzstim = np.std(psd28_40hzstim, axis=0)  
std28_30hzstim = np.std(psd28_30hzstim, axis=0)  
std28_20hzstim = np.std(psd28_20hzstim, axis=0)

ub20_40hzstim = p40_20+std20_40hzstim
ub20_30hzstim = p30_20+std20_30hzstim
ub20_20hzstim = p20_20+std20_20hzstim

ub24_40hzstim = p40_24+std24_40hzstim
ub24_30hzstim = p30_24+std24_30hzstim
ub24_20hzstim = p20_24+std24_20hzstim

ub28_40hzstim = p40_28+std28_40hzstim
ub28_30hzstim = p30_28+std28_30hzstim
ub28_20hzstim = p20_28+std28_20hzstim

lb20_40hzstim = p40_20-std20_40hzstim
lb20_30hzstim = p30_20-std20_30hzstim
lb20_20hzstim = p20_20-std20_20hzstim

lb24_40hzstim = p40_24-std24_40hzstim
lb24_30hzstim = p30_24-std24_30hzstim
lb24_20hzstim = p20_24-std24_20hzstim

lb28_40hzstim = p40_28-std28_40hzstim
lb28_30hzstim = p30_28-std28_30hzstim
lb28_20hzstim = p20_28-std28_20hzstim
```

```python
xlabel = 'Frequency (Hz)'
ylabel = 'Power Spectral Density'
savefig = True
filename = '../Figures/PSD40_ChC10_evolution_BC_time'
fontsizes = [25,35,25]

plotPSDs(freqs*1000,p40_20,p40_24,p40_28,xlabel,ylabel,savefig,filename,fontsizes)
```
